```python
from sklearn.impute import KNNImputer
impute=KNNImputer()
```

```python
for i in data.select_dtypes(include='number'):
  data[i]=impute.fit_transform(data[[i]])
```

```python
data.isnull().sum()
```

```python
for i in ['OperatingHours','StoreLocation']:
  data.fillna(data[i].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
laber_encoder=LabelEncoder()
data[['OperatingHours','StoreLocation']]=data[['OperatingHours','StoreLocation']].apply(laber_encoder.fit_transform)
```
